```python
#Preprocessing on Dataset-1
data={'rollno no':[101,102,103,104,105],'name':['raghav','rohan','neha','anik','arjit'],
      'age':[20,'-',21,22,21],'marks':[85,95,'-',88,76],'department':['CSE','IT','cse','IT','ENTC'],
      'admission_date':['01-06-2025','03-06-2025','04-06-2025','05-06-2025','06/06/2025']}
df=pd.DataFrame(data)
df
#saving as csv
df.to_csv('student_data.csv',index=False)
```

```python
#step2: Replace '-' with NAN to mark missing values
df.replace('-',np.nan,inplace=True)
#convert columns to numeric where needed
df['age']=pd.to_numeric(df['age'],errors='coerce')
df['marks']=pd.to_numeric(df['marks'],errors='coerce')
df
```

```python
#step3: check missing values
print("\nMissing Values")
print(df.isna().sum())
```

```python
#Fill missing age with mean
df['age'].fillna(df['age'].mean(),inplace=True)
#Fill missing marks with median
df['marks'].fillna(df['marks'].median(),inplace=True)
df
```

```python
#step5: Fix inconsistent text formatting
df['department'] = df['department'].str.upper()
df['department']
```

```python
#step6: Convert Admission date to standard datetime format
df['admission_date'] = pd.to_datetime(df['admission_date'],errors='coerce')
df['admission_date']
```

```python
#step7: Save cleaned dataset
df.to_csv("cleaned_student_data.csv",index=False)
```

```python
#Upload the Dataset
df=pd.read_csv('/content/Cars93.csv')
df
```

```python
df.isna().sum() #printing missing values
```

```python
#imputing missing values
#1 Luggage Room
df['Luggage.room']=df['Luggage.room'].fillna(value=df['Luggage.room'].mean())
df.isna().sum()
```

```python
#2Rear seats
df['Rear.seat.room']=df['Rear.seat.room'].fillna(value=df['Rear.seat.room'].mean())
df.isna().sum()
```

```python
#3AirBags
df['AirBags']=df['AirBags'].fillna(df['AirBags'].mode().values[0])
df.isna().sum()
```

```python
#save cleaned dataset
df.to_csv('cleaned_cars_data.csv',index=False)
```
